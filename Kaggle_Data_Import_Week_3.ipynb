```python
import opendatasets as od
download_url = 'https://www.kaggle.com/datasets/steverusso/cincinnati-car-crash-data'
od.download(download_url)
```

```python
import pandas as pd
df = pd.read_csv('./cincinnati-car-crash-data/cincinnati_traffic_crash_data__cpd.csv')
```

```python
df.head(10)
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna(subset=['INJURIES'])
```

```python
print(df.isnull().sum())
```

```python
df.head(10)
```

```python
df['CRASHSEVERITY'] = df['CRASHSEVERITY'].str[0].astype(int)
```

```python
print(df.dtypes)
```

```python
df.head(10)
```

```python
df['CRASH_YEAR'] = pd.DatetimeIndex(df['CRASHDATE']).year
```
